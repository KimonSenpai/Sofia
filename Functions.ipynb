$a^{2k} = (a^2)^k$

$a^{2k + 1} = (a^2)^k*a$

```python
def fact(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

def fact2(n):
    if 0 <= n <= 1: return 1
    
    return n*fact2(n-1)

def power(a, n):
    if n == 1:
        return a
    if n == 0:
        return 1
    
    if n%2 == 0:
        return power(a*a, n//2)
    else:
        return power(a*a, n//2)*a

print(fact2(7))
print(power(3, 8))
```

```python
def Even(lst):
    res = []
    for val in lst:
        if val%2 == 0:
            res += [val]
    return res

print(*Even([1, 2, 3, 4, 5, 6, 7]))
    
```

```python
def print_document(pages):
    for page in pages:
        if 'Секретно' in page:
            print('Дальнейшие материалы засекречены')
            return
        print(page)
    print('Напечатано без купюр')
    
print_document(['Обычная страница', 'И еще страница', 'Секретно, этот текст не показывать'])
print()
print_document(['Пустой треп','который никому','не интересен'])
```

$R_1 = (x_1, y_1, z_1)$

$R_2 = (x_2, y_2, z_2)$

$(R_1, R_2) = x_1x_2 + y_1y_2 + z_1z_2$

```python
R1 = [3, 5, 9]
R2 = [4, 7, 2]

print(sum(map(lambda x: x[0]*x[1], zip(R1, R2))))

S = []
for x1, x2 in zip(R1, R2):
    S += [x1+x2]
    
print(S)
```
