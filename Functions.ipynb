$a^{2k} = (a^2)^k$

$a^{2k + 1} = (a^2)^k*a$

```python
def fact(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

def fact2(n):
    if 0 <= n <= 1: return 1
    
    return n*fact2(n-1)

def power(a, n):
    if n == 1:
        return a
    if n == 0:
        return 1
    
    if n%2 == 0:
        return power(a*a, n//2)
    else:
        return power(a*a, n//2)*a

print(fact2(7))
print(power(3, 8))
```

```python
from sys import stdin

max_length = 0
max_correct = [-1]
for line in stdin:
 
    numbers = line.split(':')
    correct = []
    for number in numbers:
        flag = True
        for dig in number[-1::-2]:
            flag = flag and (int(dig) % 2 == 0)
        flag = flag and (sum([int(dig) for dig in number]) % 2 == 1)
        if flag:
            correct.append(number)
    if len(correct) > max_length:
        max_length = len(correct)
        max_correct = correct
        
print(*max_correct)
```

```python
def shift(s, **arg):
    if 'step' in arg:
        step = arg['step'] % len(s)
   
        return s[len(s) - step:len(s)] + s[0:len(s) - step]

print(shift('Hello', step=2))
```
