```python
def Number(n):
    Dig = ('ноль', 'один', 'два', 'три', 'четыре', 'девятнадцать')
    Dig2 = ('двадцать', 'тридцать')
    if n < 20:
        return Dig[n]
    elif n%10 == 0:
        return Dig2[n//10 - 2]
    else:
        return Dig2[n//10 - 2] + ' ' + Dig[n%10]
    
print(Number(4))
print(Number(30))
print(Number(24))
```

$a^{2k} = (a^2)^k$

$a^{2k + 1} = (a^2)^k*a$

```python
def fact(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

def fact2(n):
    if 0 <= n <= 1: return 1
    
    return n*fact2(n-1)

def power(a, n):
    if n == 1:
        return a
    if n == 0:
        return 1
    
    if n%2 == 0:
        return power(a*a, n//2)
    else:
        return power(a*a, n//2)*a

print(fact2(7))
print(power(3, 8))
```
