## Последовательности

```python
count = 0
while True:
    val = int(input())
    
    if val == 0: break
    
    if val%2 == 0: count += 1

print(count)
```

```python
count = 0
summ = 0
while True:
    value = int(input())
    
    if value == 0: break
        
    count += 1
    summ += value
    
print(summ/count)
```

```python
Max = 0
while True:
    value = int(input())
    
    if value == 0: break
        
    if Max < value: Max = value
    
print(Max)
```

```python
count = 0
prev = int(input())

while prev != 0:
    val = int(input())
    if val == 0:
        break
    if prev < val:
        count += 1
    prev = val
print(count)
```

```python
max1, max2 = 0, 0
while True:
    val = int(input())
    if val == 0: break;
    
    if val > max1:
        max2 = max1
        max1 = val
    elif val > max2:
        max2 = val
        
print(max2)
```

```python
maxcount = 1
count = 1
val2 = int(input())
while True:
    val = int(input())
    if val == 0:
        break
    
    if val2 != val:
        count = 1
    else:
        count += 1
    if maxcount < count:
        maxcount = count
    val2 = val
print(maxcount)
```

```python
num = int(input())
s = 0
while num > 0:
    s += num % 10
    num //= 10
print(s)
```
