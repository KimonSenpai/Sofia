```python
d = docx.Document()
import sys
import importlib as il
import setdefaultencoding
#il.reload(sys)
#setdefaultencoding.setdefaultencoding('cp1251')
p = 0
f = open("input.txt", 'r', encoding='utf-8')
place = f.readline()
time = f.readline()
for s in f:
    d.add_heading("Приглашение").alignment = docx.enum.text.WD_ALIGN_PARAGRAPH.CENTER
    p = d.add_paragraph(F"Приглашаю вас, {s} на встречу {place} {time}! ")
    p.add_run("Очень ждем!").bold = True
    p.runs[1].add_break(docx.enum.text.WD_BREAK.PAGE)
d.save("sample.docx")
f.close()
```
