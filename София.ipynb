## Палиндромы

$i = \overline{i_{n-1}i_{n-2}i_{n-3}...i_2i_1i_0}$

$i = 123$

$i_0 = 3, i_1 = 2, i_2 = 1$

$i_k = i//10^k\%10$

```python
def Dig(n, i):
    return n//10**i%10
def Abs(x):
    if x >= 0:
        return x
    else:
        return -x

k = int(input())
count = 0
#y = Abs(x)
for i in range(1, k+1):
    l = 0
    t = i
    while t != 0:
        t //= 10
        l += 1
    #f := число является палиндромом
    f = True
    for j in range(l//2):
        if Dig(i, j) != Dig(i, l-1 - j):
            f = False
    if f:
        count += 1
        
print(count)
```

# Numeral systems

$\Sigma = \{0, 1\}$ -- алфавит

$q = 2$ -- основание

$\{q^0, q^1, q^2, ...\}$ -- базис

$a = \overline{a_{n-1}...a_1a_0}_q = a_0q^0 + a_1q^1 +...+ a_{n-1}q^{n-1}$

$123 = 100 + 20 + 3 = 3\cdot 10^0 + 2\cdot 10^1 + 1\cdot 10^2$

$a\%q = a_0\cdot q^0 = a_0$

$a//q = a_1q^0 + a_2q^1+...+ a_{n-1}q^{n-2}$

$a//q\%q = a_1\cdot q^0 = a_1$

$a//q^2 = a_2q^0 + a_3q^1+...+ a_{n-1}q^{n-3}$

$a//q^2\%q = a_2\cdot q^0 = a_2$

```python
n = int(input())
while n > 0:
    print(n%2, end='')
    n //= 2
print()
```

## Максимальная длина монотонного фрагмента

```python
def main():
    MaxLen = 1
    prev = int(input())
    if prev == 0:
        print(0)
        return
    Len = 0
    Ind = 0
    while True:
        val = int(input())
        if val == 0:
            break
        if Ind == 1 and val > prev or \
            Ind == -1 and val < prev:
                Len += 1
        elif val == prev:
            MaxLen = max(MaxLen, Len)#if Len > MaxLen: MaxLen = Len
            Len = 1
            Ind = 0
        elif val > prev:
            MaxLen = max(MaxLen, Len)
            Len = 2
            Ind = 1
        else:
            MaxLen = max(MaxLen, Len)
            Len = 2
            Ind = -1
        prev = val
    MaxLen = max(MaxLen, Len)
    print(MaxLen)
main()

```
